# This notebook makes predictions for all of the "Full" Linear Regression csv's. 
### These include:
- Full Linear Regression --> Using All Years and Counties to predict one missing point
- Full Haversine IDW --> Same as above, but with IDW weights of (1/x)
- Full Haversine IDW (Power of 2) --> (1/x**2)

```python
from os.path import join 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import rbf_kernel
from sklearn.neighbors import DistanceMetric
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import statsmodels.api as sm
import statsmodels.formula.api as smf
from math import pi
import numpy as np

%matplotlib inline
plt.style.use('seaborn')
pd.set_option('display.max_columns', 99)
```

```python
data_dir = "../data/"
target_crop = "soybean"
full_data = pd.read_csv(join(data_dir, "{}_full_centroids.csv".format(target_crop)))
full_data.fips_code = full_data.fips_code.astype(str).apply(lambda x: x.zfill(5))
full_data.set_index('fips_code', inplace=True)
full_data.head()
```

```python
year_cols = list(map(str, range(1970, 2018)))
X = full_data[year_cols]
counties, years = X.shape
X.head()
```

# Predict for only 2017, using all other years data

```python
X_train = X.copy()
y_train = X_train.pop("2017")

preds = []

for i in range(counties):
    x = X_train[:i].append(X_train[i+1:])
    y = y_train[:i].append(y_train[i+1:])
    lm = LinearRegression()
    lm.fit(x, y)
    y_pred = lm.predict(X_train[i:i+1])
    preds.append(y_pred[0])
```

```python
print("RMSE: {:.4f}".format(np.sqrt(mean_squared_error(y_train, preds))))
print("R2: {:.4f}".format(r2_score(y_train, preds)))
```

# Repeat for all years

```python
lr_years = {}

for year in X.columns.values:
    
    X_train = X.copy()
    y_train = X_train.pop(year)
    preds = []

    for i in range(X.shape[0]):
        x = X_train[:i].append(X_train[i+1:])
        y = y_train[:i].append(y_train[i+1:])
        lm = LinearRegression()
        lm.fit(x, y)
        y_pred = lm.predict(X_train[i:i+1])
        preds.append(y_pred[0])
        
    lr_years[year] = preds
```

```python
lin_reg_preds_df = pd.DataFrame(lr_years)
lin_reg_preds_df.index = full_data.index

lin_reg_preds_df.head()
```

### Save results for Multivariate Linear Regression (all years to predict the missing year)

```python
lin_reg_preds_df.to_csv(join(data_dir, "{}_lin_reg_preds.csv".format(target_crop)))
```

# Metrics for Linear Regression

```python
def mape(y_pred,y_true):
    y_pred = np.asarray(y_pred)
    y_true = np.asarray(y_true)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
lin_reg_RMSE = np.sqrt(mean_squared_error(X, lin_reg_preds_df))
lin_reg_R2 = r2_score(X, lin_reg_preds_df)
lin_reg_MAPE = mape(X, lin_reg_preds_df)
lin_reg_MAE = mean_absolute_error(X, lin_reg_preds_df)

print("Linear Regression RMSE: {:.4f}".format(lin_reg_RMSE))
print("Linear Regression R2: {:.4f}".format(lin_reg_R2))
print("Linear Regression MAPE: {:.4f}".format(lin_reg_MAPE))
print("Linear Regression MAE: {:.4f}".format(lin_reg_MAE))
```

# By year

```python
lin_reg_r2 = {}
lin_reg_rmse = {}
lin_reg_mape = {}
lin_reg_mae = {}

for i in lin_reg_preds_df.columns:
    lin_reg_rmse[i] = np.sqrt(mean_squared_error(X[i], lin_reg_preds_df[i]))
    lin_reg_r2[i] = r2_score(X[i], lin_reg_preds_df[i])
    lin_reg_mape[i] = mape(X[i], lin_reg_preds_df[i])
    lin_reg_mae[i] = mean_absolute_error(X[i], lin_reg_preds_df[i])
```

```python
lin_reg_metrics_df = pd.DataFrame({"RMSE": lin_reg_rmse, "R2": lin_reg_r2,
                                  "MAPE": lin_reg_mape, "MAE": lin_reg_mae})
lin_reg_metrics_df.head()
```

```python
lin_reg_metrics_df[['RMSE', 'MAE']].plot(figsize=(12,8), rot=90)
plt.ylabel("Error {} bushels/acre)".format(target_crop), fontsize=18)
plt.xlabel("Year", fontsize=18)
plt.xticks(range(years), range(1970, 2018));
```

# Weighted LR (Inverse Haversine Distance)

```python
full_spatial = full_data[['latitude', 'longitude']]

full_hav_df = full_spatial.copy()
full_hav_df = np.radians(full_hav_df)
full_hav_df.head()
```

```python
dist = DistanceMetric.get_metric("haversine")
hav_dist_df = pd.DataFrame(dist.pairwise(full_hav_df))
```

```python
# Predict for only 2017
X_train = X.copy()
y_train = X_train.pop("2017")
hav_weights = hav_dist_df.copy()
hav_weights = hav_weights.applymap(lambda x: x**-1 if x != 0 else 0)
hav_weights = hav_weights.apply(lambda x: x/sum(x), axis=1) # normalize
hav_preds = []

for i in range(X.shape[0]):
    x = X_train[:i].append(X_train[i+1:])
    y = y_train[:i].append(y_train[i+1:])
    w = hav_weights[i][:i].append(hav_weights[i][i+1:])
    lm = LinearRegression()
    lm.fit(x, y, w)
    y_pred = lm.predict(X_train[i:i+1])
    hav_preds.append(y_pred[0])
```

### Best yet

```python
print("RMSE: {:.4f}".format(np.sqrt(mean_squared_error(y_train, hav_preds))))
print("R2: {:.4f}".format(r2_score(y_train, hav_preds)))
```

```python
hav_lr_years = {}
hav_weights = hav_dist_df.copy()
hav_weights = hav_weights.applymap(lambda x: x**-1 if x != 0 else 0)
hav_weights = hav_weights.apply(lambda x: x/sum(x), axis=1) # normalize

for year in X.columns.values:
    
    X_train = X.copy()
    y_train = X_train.pop(year)
    hav_preds = []

    for i in range(X.shape[0]):
        x = X_train[:i].append(X_train[i+1:])
        y = y_train[:i].append(y_train[i+1:])
        w = hav_weights[i][:i].append(hav_weights[i][i+1:])
        lm = LinearRegression()
        lm.fit(x, y, w)
        y_pred = lm.predict(X_train[i:i+1])
        hav_preds.append(y_pred[0])
        
    hav_lr_years[year] = hav_preds
```

```python
hav_lr_preds_df = pd.DataFrame(hav_lr_years)
hav_lr_preds_df.index = full_data.index

hav_lr_preds_df.to_csv(join(data_dir, "{}_lr_hav_idw_preds.csv".format(target_crop)))
```

```python
hav_lr_RMSE = np.sqrt(mean_squared_error(X, hav_lr_preds_df))
hav_lr_R2 = r2_score(X, hav_lr_preds_df)
hav_lr_MAPE = mape(X, hav_lr_preds_df)
hav_lr_MAE = mean_absolute_error(X, hav_lr_preds_df)

print("Haversine Inverse Distance Weighted Linear Regression RMSE: {:.4f}".format(hav_lr_RMSE))
print("Haversine Inverse Distance Weighted Linear Regression R2: {:.4f}".format(hav_lr_R2))
print("Haversine Inverse Distance Weighted Linear Regression MAPE: {:.4f}".format(hav_lr_MAPE))
print("Haversine Inverse Distance Weighted Linear Regression MAE: {:.4f}".format(hav_lr_MAE))
```
