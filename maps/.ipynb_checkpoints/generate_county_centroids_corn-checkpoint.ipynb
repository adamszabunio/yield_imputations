```python
from pyspatial.vector import read_layer
import pandas as pd
pd.set_option("display.max_columns", 99)
```

```python
ts_df = pd.read_csv("../data/ts_corn_df.csv")
ts_df.fips_code = ts_df.fips_code.astype(str).apply(lambda x: x.zfill(5))

ts_df.head()
```

```python
url = "http://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_county_500k.zip"
vl, vldf = read_layer(url, index="GEOID")
vl.head()
```

```python
vldf.head()
```

```python
centroids = vl.centroids(format='DataFrame').reset_index()
centroids.rename(columns={'index':'fips_code', 'x':'longitude', 'y':'latitude'}, inplace=True)

centroids.head()
```

```python
# centroids[centroids.fips_code.isin(['29193', '46131'])] # fips code clashes
centroids[centroids.fips_code.isin(['29186', '46071'])]
```

```python
ts_df.shape, centroids.shape
```

```python
ts_centroids_df = ts_df.merge(centroids)
print(ts_centroids_df.shape)
ts_centroids_df.head()
```

```python
set(ts_df.fips_code).difference(set(ts_centroids_df.fips_code)) # make sure all counties are captured 
```

```python
ts_centroids_df.to_csv("../data/ts_centroids_corn.csv")
```
