```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 99)
plt.style.use("fivethirtyeight")
```

```python
metrics_df = pd.read_csv("../data/bootstrap_corn_metrics.csv", index_col=0)
metrics_df.index = [i[:-5] for i in metrics_df.index.values]
metrics_df
```

```python
metrics_df.index
```

```python
methods_dict = {
'k_preds': [i.lower() for i in metrics_df.index.values if i.startswith('K')],
'exp_preds': [i.lower() for i in metrics_df.index.values if i.startswith('EXP')],
'full_preds': [i for i in metrics_df.index.values if i.startswith('FULL')],
'baseline': ['lr_pred', 'ffill', 'county_mean']
}
methods_dict['baseline']
```

```python
labels_dict = {
    'mae': "Corn bushels/acre",
    'mape': "Mean % error",
    'rmse': "Corn bushels/acre",
    'r2': "$r^2$ value"
    }
labels_dict["mape"]
```

```python
preds_df = pd.read_csv("../data/bootstrap_corn_predicitions.csv", index_col=0)
preds_df.index = list(map(lambda x: str(x).zfill(5), preds_df.index.tolist()))

print(preds_df.shape)
preds_df.head()
```

```python
y_true_df = preds_df[[col for col in preds_df.columns if "y_true" in col]]
y_true_df.head()
```

```python
lr_pred_df = preds_df[[col for col in preds_df.columns if "lr_pred" in col]]
lr_pred_df.head()
```

```python
from sklearn.metrics import r2_score, mean_squared_error, median_absolute_error

def fips_met(metric, y_true, y_pred):
    '''Returns a metric of choice for each fips code
    (e.g. r2score(true, pred) for 1000 bootstrap sample predictions)'''
    true, pred = np.array(y_true), np.array(y_pred)
    return list(map(lambda a, b: metric(a, b), true, pred))
```

```python
baseline_df = pd.DataFrame(index=y_true_df.index)

for met in methods_dict["baseline"]:
    pred_df = preds_df[[col for col in preds_df.columns if met in col]]
    residuals = (np.array(pred_df) - np.array(y_true_df)) # know that residuals are usually calculated y-yhat
    mean_ = np.apply_along_axis(np.mean, axis=1, arr=residuals) # choose to do it this way for ea
    median_ = np.apply_along_axis(np.median, axis=1, arr=residuals) 
    r2 = fips_met(r2_score, y_true_df, pred_df)
    mae = fips_met(median_absolute_error, y_true_df, pred_df)
    rmse = np.sqrt(fips_met(mean_squared_error, y_true_df, pred_df))
    baseline_df[met+"_MEAN"] = mean_
    baseline_df[met+"_MEDIAN"] = median_
    baseline_df[met+"_r2SCORE"] = r2
    baseline_df[met+"_MAE"] = mae
    baseline_df[met+"_RMSE"] = rmse

baseline_df.head()
```

```python
methods_dict["k_preds"][0] = "k1_pred"

k_preds_df = pd.DataFrame(index=y_true_df.index)

for met in methods_dict["k_preds"]:
    pred_df = preds_df[[col for col in preds_df.columns if met in col]]
    residuals = (np.array(pred_df) - np.array(y_true_df))
    mean_ = np.apply_along_axis(func1d=np.mean, axis=1, arr=residuals)
    median_ = np.apply_along_axis(func1d=np.median, axis=1, arr=residuals)
    r2 = fips_met(r2_score, y_true_df, pred_df)
    mae = fips_met(median_absolute_error, y_true_df, pred_df)
    rmse = np.sqrt(fips_met(mean_squared_error, y_true_df, pred_df))
    k_preds_df[met+"_MEAN"] = mean_
    k_preds_df[met+"_MEDIAN"] = median_
    k_preds_df[met+"_r2SCORE"] = r2
    k_preds_df[met+"_MAE"] = mae
    k_preds_df[met+"_RMSE"] = rmse

k_preds_df.head()
```

```python
methods_dict["exp_preds"][2] = "exp1_"
methods_dict["exp_preds"][3] = "exp2_"

exp_preds_df = pd.DataFrame(index=y_true_df.index)

for met in methods_dict["exp_preds"]:
    pred_df = preds_df[[col for col in preds_df.columns if met in col]]
    residuals = (np.array(pred_df) - np.array(y_true_df))
    mean_ = np.apply_along_axis(func1d=np.mean, axis=1, arr=residuals)
    median_ = np.apply_along_axis(func1d=np.median, axis=1, arr=residuals)
    r2 = fips_met(r2_score, y_true_df, pred_df)
    mae = fips_met(median_absolute_error, y_true_df, pred_df)
    rmse = np.sqrt(fips_met(mean_squared_error, y_true_df, pred_df))
    exp_preds_df[met+"_MEAN"] = mean_
    exp_preds_df[met+"_MEDIAN"] = median_
    exp_preds_df[met+"_r2SCORE"] = r2
    exp_preds_df[met+"_MAE"] = mae
    exp_preds_df[met+"_RMSE"] = rmse

exp_preds_df.head()
```

```python
methods_dict["full_preds"][1] = "FULL_HAV_LR_pred"
full_preds_df = pd.DataFrame(index=y_true_df.index)

for met in methods_dict["full_preds"]:
    pred_df = preds_df[[col for col in preds_df.columns if met in col]]
    residuals = (np.array(pred_df) - np.array(y_true_df))
    mean_ = np.apply_along_axis(func1d=np.mean, axis=1, arr=residuals)
    median_ = np.apply_along_axis(func1d=np.median, axis=1, arr=residuals)
    r2 = fips_met(r2_score, y_true_df, pred_df)
    mae = fips_met(median_absolute_error, y_true_df, pred_df)
    rmse = np.sqrt(fips_met(mean_squared_error, y_true_df, pred_df))
    full_preds_df[met+"_MEAN"] = mean_
    full_preds_df[met+"_MEDIAN"] = median_
    full_preds_df[met+"_r2SCORE"] = r2
    full_preds_df[met+"_MAE"] = mae
    full_preds_df[met+"_RMSE"] = rmse

full_preds_df.head()
```
