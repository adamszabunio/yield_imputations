```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
pd.set_option("display.max_columns", 99)
```

```python
metrics_df = pd.read_csv("../data/bootstrap_corn_metrics.csv", index_col=0)
metrics_df.index = [i[:-5] for i in metrics_df.index.values]
metrics_df
```

```python
metrics_df.index
```

```python
methods_dict = {
'k_preds': [i for i in metrics_df.index.values if i.startswith('K')],
'exp_preds': [i for i in metrics_df.index.values if i.startswith('EXP')],
'full_preds': [i for i in metrics_df.index.values if i.startswith('FULL')],
'baseline': ['LR', 'FFILL', 'COUNTY_MEAN']
}
methods_dict['baseline']
```

```python
metrics_dict = {
    'mae': [col for col in metrics_df.columns if 'mae' in col.lower()],
    'mape': [col for col in metrics_df.columns if 'mape' in col.lower()],
    'rmse': [col for col in metrics_df.columns if 'rmse' in col.lower()],
    'r2': [col for col in metrics_df.columns if 'rsqu' in col.lower()]
    }
metrics_dict["mape"][:5]
```

```python
labels_dict = {
    'mae': "Corn bushels/acre",
    'mape': "Mean % error",
    'rmse': "Corn bushels/acre",
    'r2': "$r^2$ value"
    }
labels_dict["mape"]
```

```python
fig, ax = plt.subplots(figsize=(13,8))
plt.title("1000 Sample Bootstrap Mean For: MAE metric (w/ STD ERR)")
plt.xlabel(labels_dict['mae'])
mae_df = metrics_df[metrics_dict['mae']]
mae_df.agg('mean', axis=1).plot.barh(xerr=mae_df.agg('std', axis=1), error_kw={"capsize":2,"markeredgewidth":1, "alpha":.6});
fig.savefig('../images/all_bootstrap_preds_mae')
```

```python
fig, ax = plt.subplots(figsize=(13,8))
plt.title("1000 Sample Bootstrap Mean For: MAPE metric (w/ STD ERR)")
plt.xlabel(labels_dict['mape'])
mape_df = metrics_df[metrics_dict['mape']]
mape_df.agg('mean', axis=1).plot.barh(xerr=mape_df.agg('std', axis=1), error_kw={"capsize":2,"markeredgewidth":1, "alpha":.6});
fig.savefig('../images/all_bootstrap_preds_mape')
```

```python
fig, ax = plt.subplots(figsize=(13,8))
plt.title("1000 Sample Bootstrap Mean For: RMSE metric (w/ STD ERR)")
plt.xlabel(labels_dict['rmse'])
rmse_df = metrics_df[metrics_dict['rmse']]
rmse_df.agg('mean', axis=1).plot.barh(xerr=rmse_df.agg('std', axis=1), error_kw={"capsize":2,"markeredgewidth":1, "alpha":.6});
fig.savefig('../images/all_bootstrap_preds_rmse')
```

```python
fig, ax = plt.subplots(figsize=(13,8))
plt.title("1000 Sample Bootstrap Mean For: $r^2$ metric (w/ STD ERR)")
plt.xlabel(labels_dict['r2'])
r2_df = metrics_df[metrics_dict['r2']]
r2_df.agg('mean', axis=1).plot.barh(xerr=r2_df.agg('std', axis=1), error_kw={"capsize":2,"markeredgewidth":1, "alpha":.6});
fig.savefig('../images/all_bootstrap_preds_r2')
```

```python
def mean_metric(metrics_df, method, metric, fig_name=None, save_dir='../images/'):
    '''
    creates horizontal bar chart of mean metric, w/ std_err bars
    inputs:
    dataframe: metric_df
    methods: str() methods to include (e.g. 'k_preds')
    metric: str() all columns w/ metric of choice (e.g. 'rmse')
    fig_name: str(name_of_figure)
    returns: None
    plots horizontal barchart of the above
    optionally saves figure to default 'images/' dir'''
    
    metric_cols = metrics_dict[metric]
    method_idx = methods_dict[method]
    fig, ax = plt.subplots(figsize=(13,8))
    plt.title(method.upper() + " 1000 Sample Bootstrap Mean \n For: " + \
              metric.upper() + " metric (w/ STD ERR)")
    plt.xlabel(labels_dict[metric])
    met_df = metrics_df.loc[method_idx][metric_cols]
    met_df.agg('mean', axis=1).plot.barh(xerr=met_df.agg('std', axis=1), \
                                           error_kw={"capsize":2,\
                                                     "markeredgewidth":1, \
                                                     "alpha":.6});
    if fig_name:
        fig.savefig(save_dir+fig_name)
```

```python
mean_metric(metrics_df, 'k_preds', 'mape', 'k_preds_bootstrap_mape')
```

```python
mean_metric(metrics_df, 'k_preds', 'r2', 'k_preds_bootstrap_r2')
```

```python
mean_metric(metrics_df, 'k_preds', 'rmse', 'k_preds_bootstrap_rmse')
```
